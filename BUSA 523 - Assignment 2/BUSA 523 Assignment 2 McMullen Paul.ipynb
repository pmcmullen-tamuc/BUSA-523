# Question 1.
- Write a program for flipping a coin 10,000 times and store the results in a list. After which, identify the number of streaks. (streak - a series of 5 or more heads or tails)

```python
import random

resultList = []
currentStreak = 1
previousResult = ""
streak = 0
streakHeads = 0
streakTails = 0
maxStreak = 1

for i in range(10000):
    randomNumber = random.randint(0, 1)
    
    if randomNumber == 0:
        result = "heads"
    else:
        result = "tails"
        
    resultList.append(result)

for data in resultList:
    if data == previousResult:
        currentStreak += 1
        
        if currentStreak > maxStreak:
            maxStreak = currentStreak
            
    else:
        if currentStreak >= 5:
            if previousResult == "heads":
                streakHeads += 1
            else:
                streakTails +=1
                
            streak += 1
        
        currentStreak = 1
        
    previousResult = data

    
if currentStreak >= 5:
    if previousResult == "heads":
        streakHeads += 1
    else:
        streakTails += 1
        
    streak += 1
    

print("Total number of streaks of 5 or more: " + str(streak))
print("Number of streaks of heads: " + str(streakHeads))
print("Number of streaks of tails: " + str(streakTails))
print("The highest number of consecutive results was: " + str(maxStreak))
```

# Question 2.
- Write a program to take user inputs (number of swords, diamonds, gold coins, ropes, and potions) for a video game and store them in a dictionary. After which print "Inventory: " followed by the number of items and name of each category.

```python
userCache = {'swords': 0, 'diamonds': 0, 'gold': 0, 'coins': 0, 'ropes': 0, 'potions': 0}

errorCount = 0

for key in userCache.keys():
    
    if errorCount > 1:
        print("Sorry, you've entered too many incorrect values. Please try again later.")
        break
    else:
        errorCount = 0
            
    print("Please enter the number of " + key + ":")
    
    for i in range(2):
        try:
            userInput = int(input())
            if userInput >= 0:
                userCache[key] = userInput
                break
            else:
                print("Please enter a nonnegative integer value for " + key + ":")
                errorCount += 1
            
        except:
            print("Please enter a nonnegative integer value for " + key + ":")
            errorCount +=1
        
if errorCount < 2:
    print("")
    print("Inventory:")
    for key, value in userCache.items():
        print(str(value) + " " + key)
```

# Question 3.
- Repeat question 1 using arrays.

```python
import random
import numpy as np

currentStreak = 1
previousResult = -1
streak = 0
streakHeads = 0
streakTails = 0
maxStreak = 1


# heads: 0
# tails: 1
randomNum = np.random.randint(2, size = (10000, 1))

for row in range(len(randomNum)):
    if randomNum[row][0] == previousResult:
        currentStreak += 1
        
        if currentStreak > maxStreak:
            maxStreak = currentStreak
            
    else:
        if currentStreak >= 5:
            if previousResult == 0:
                streakHeads += 1
            else:
                streakTails +=1
                
            streak += 1
        
        currentStreak = 1
        
    previousResult = randomNum[row][0]

    
if currentStreak >= 5:
    if previousResult == 0:
        streakHeads += 1
    else:
        streakTails += 1
        
    streak += 1
    

print("Total number of streaks of 5 or more: " + str(streak))
print("Number of streaks of heads: " + str(streakHeads))
print("Number of streaks of tails: " + str(streakTails))
print("The highest number of consecutive results was: " + str(maxStreak))

```

# Question 4.
- Create a game with the following instrctions:
    - a. There are 3 players and 10 iterations.
    - b. In each iteration, every player rolls a die.
    - c. The winner of the game is the one who has the highest score when rolls from all the iterations are added.

```python
import random
import numpy as np

results = []

for rounds in range(1, 11):
    results = []
    for player in range(1, 4):
        diceRoll = random.randint(1, 6)
        results.append(diceRoll)
    
    if rounds == 1:
        allResults = np.array([results])
    else:
        allResults = np.append(allResults, np.array([results]), axis = 0)
        

player1, player2, player3 = sum(allResults)

if player1 > player2 and player1 > player3:
    print("Congratulations Player 1")
elif player2 > player1 and player2 > player3:
    print("Congratulations Player 2")
elif player3 > player1 and player3 > player2:
    print("Congratulations Player 3")
elif player1 == player2 and player1 == player3:
    print("Three way tie!!! What are the odds?")
elif player1 == player2:
    print("Player 1 and Player 2 have tied")
elif player1 == player3:
    print("Player 1 and Player 3 have tied")
elif player2 == player3:
    print("Player 2 and Player 3 have tied")
else:
    print("Unexpected result. Well this is rather embarrassing!")
    
print("")
print("Player 1: " + str(player1))
print("Player 2: " + str(player2))
print("Player 3: " + str(player3))
```
