# Question 1

- Write a program which randomly selects a number between 1 and 20. The user is then given 5 chances to guess the number correctly to win.

```python
import random

randomNumber = random.randint(1, 20)
correct = 0

for i in range(5):
    
    remainingGuessesMsg = "You have " + str(4-i) + " guesses remaining."
    
    print("Please guess an interger between 1 and 20.")
    
    try:
        userInput = int(input())
        if 1 <= userInput <= 20:
            if userInput == randomNumber:
                print("Congratulations, you won. You must have cheated!")
                correct = 1
                break
            else:
                print("Sorry, that is incorrect. " + remainingGuessesMsg)
        else:
            print("Your guess must be between 1 and 20. " + remainingGuessesMsg)
    except:
        print("Your must enter an interger value. " + remainingGuessesMsg)
        
    print("")
    
if correct == 0:
    print("The correct answer was " + str(randomNumber) + '.')
    
```

# Question 2

- After each guess, print if the next guess should be lower or higher.

# Question 3.
- Write a program for "Rock-Paper-Scissors-Lizard-Spock"

```python
import random

print("Enter your choice: Rock, Paper, Scissors, Lizard, or Spock")

userInput = input().lower()

if userInput not in ["rock", "paper", "scissors", "lizard", "spock"]:
    print("Sorry, that is not a valid response.")
else:
    randomNumber = random.randint(1, 5)
    
    if randomNumber == 1:
        computerResponse = "rock"
    elif randomNumber == 2:
        computerResponse = "paper"
    elif randomNumber == 3:
        computerResponse = "scissors"
    elif randomNumber == 4:
        computerResponse = "lizard"
    else:
        computerResponse = "spock"
       
    print("")
    print("Computer chose: " + computerResponse)
    print("")
    
    if userInput == computerResponse:
        print("It's a tie: " + userInput + " = " + computerResponse)
    
    elif userInput == "rock":
        if computerResponse == "paper":
            print("You lose: paper covers rock!")
        elif computerResponse in ["lizard", "scissors"]:
            print("You win: rock crushes " + computerResponse + "!")
        else:
            print("You lose: Spock vaporizes rock!")
    
    elif userInput == "paper":
        if computerResponse == "rock":
            print("You win: paper covers rock!")
        elif computerResponse == "scissors":
            print("You lose: scissors cuts paper!")
        elif computerResponse == "lizard":
            print("You lose: lizard eats paper!")
        else:
            print("You win: paper disproves Spock!")
    
    elif userInput == "scissors":
        if computerResponse == "rock":
            print("You lose: rock crushes scissors!")
        elif computerResponse == "paper":
            print("You win: scissors cuts paper!")
        elif computerResponse == "lizard":
            print("You win: scissors decapitates lizard!")
        else:
            print("You lose: Spock smashes scissors!")
    
    elif userInput == "lizard":
        if computerResponse == "rock":
            print("You lose: rock crushes lizard!")
        elif computerResponse == "paper":
            print("You win: lizard eats paper!")
        elif computerResponse == "scissors":
            print("You lose: scissors decapitates lizard!")
        else:
            print("You win: lizard poisons Spock!")
    
    else:
        if computerResponse == "rock":
            print("You win: Spock vaporizes rock!")
        elif computerResponse == "paper":
            print("You lose: paper disproves Spock!")
        elif computerResponse == "scissors":
            print("You win: Spock smashes scissors!")
        else:
            print("You lose: lizard poisons Spock!")
    
```
