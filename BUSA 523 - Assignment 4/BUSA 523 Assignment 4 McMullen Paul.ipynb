# A. Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-muted")

data = pd.read_csv("data.csv", header = 0, usecols = ['dept_1 sales', 'dept_2 sales', 'dept_3 sales'])

data.plot.hist(bins = 10, title = 'Histogram of Weekly Sales')
```

# B. Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-muted")

data = pd.read_csv("data.csv", header = 0, usecols = ['dept_1 sales', 'dept_2 sales', 'dept_3 sales'])
data.agg(['sum']).plot.bar(title = 'Total Sales by Department', xlabel = 'Department', ylabel = '$ Million')
data.agg(['sum'])
```

# C. Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-muted')

data = pd.read_csv('data.csv', header = 0, usecols = ['Date', 'dept_1 sales', 'dept_2 sales', 'dept_3 sales'])

data_2011 = data.loc[(data['Date'].str[-2:] == '11')]

data_2011.plot.line(x = 'Date', ylabel = '$ Sales', title = 'Sales by Week - Calendar Year 2011')
```

# D. Area Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-muted")

data = pd.read_csv("data.csv", header = 0, usecols = ['Date', 'dept_1 sales', 'dept_2 sales', 'dept_3 sales'])

data.plot.area(x = 'Date', ylabel = '$ Sales', title = 'Departmental Sales')
```

# E. Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn-muted")

data = pd.read_csv('data.csv', header = 0, usecols = ['Date', 'Temperature', 'Fuel_Price'])
# data_2011 = data.loc[(data['Date'].str[-2:] > '10') & (data['Date'].str[-2:] < '12')]
data_2011 = data.loc[(data['Date'].str[-2:] == '11')]

data_2011.plot.scatter(x = 'Temperature', y = 'Fuel_Price', ylabel = 'Fuel Price', title = 'Scatter Plot Fuel Price vs. Temperature (2011)')
```
