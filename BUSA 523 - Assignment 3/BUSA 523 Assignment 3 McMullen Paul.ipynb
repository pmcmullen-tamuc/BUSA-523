# Question 1.
- Create a class to store student records. Each student has a first name, last name, country, overall grade (0 - 4.0 scale), credits completed (total 30) and major.

- Create a method within this class to evaluate student progress.
    - a. Progress - "Good": if grades are above 3.0 and credits completed is greater than 20
    - b. Progress - "Average": if grades are above 2.5 or credits completed is between 10 and 20
    - c. Progress - "Bad": if grades are below 2.5

```python
class studentRecord:
    def __init__(self, firstName, lastName, country, overallGrade, creditsCompleted, major):
        self.firstName = firstName
        self.lastName = lastName
        self.country = country
        self.overallGrade = overallGrade
        self.creditsCompleted = creditsCompleted
        self.major = major

    def studentProgress(self):
        if self.overallGrade > 3.0 and self.creditsCompleted > 20:
            progress = "Good"
        elif self.overallGrade > 2.5 or (10 <= self.creditsCompleted <= 20):
            progress = "Average"
        elif self.overallGrade < 2.5:
            progress = "Bad"
        else:
            progress = "Unknown"
            
        print("Student " + self.firstName + " " + self.lastName + " - Progress: " + progress)
        
        
student_1 = studentRecord('John', 'Doe', 'USA', 2.4, 21, 'BUSA')
student_1.studentProgress()
```

# Question 2.
- Write a function which separates the positive and negative numbers from a list of numbers.
    - Input: [-6,5,-3,-2,1,0,-8,9,3]
    - Output: [-6,-3,-2,-8],[5,1,9,3]

```python
def positivesAndNegatives(inputList):
    positiveList = []
    negativeList = []
    
    for data in inputList:
        if data < 0:
            negativeList.append(data)
        elif data > 0:
            positiveList.append(data)
            
    print((negativeList, positiveList))
    
    
positivesAndNegatives([-6, 5, -3, -2, 1, 0, -8, 9, 3])
```

# Question 3.
- Write a function which otputs all winning possibilities for a given player's score. Assume the score is always between 15 and 21. Further, the dealer is dealt only 2 cards.
    - Input - 18 (player's final score)
    - Output - [(9,10),(10, 11),(10,10),(11,9),(11,8)]
    
    Notes:
        - No need to specify face cards separately, you can use the value "10"
        - (11,11) is not a winning combination

```python
import random

def winningHands(playersScore):        
    if 15 <= playersScore <= 21:
        winningHand = []
        for winningScore in range(playersScore + 1, 22):
            for firstCard in range(winningScore - 11, 11):
                if firstCard > winningScore - firstCard:
                    break
                else:
                    winningHand.append((firstCard, winningScore - firstCard))
                        
        print("Player's score: " + str(playersScore))
        print("Winning hands: " + str(winningHand))
        
    else:
        print("Invalid player's score")
    
    
winningHands(random.randint(15, 21))
```
